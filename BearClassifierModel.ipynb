```python
bear_types = {
    'grizzly': grizzly_bear_imgs,
    'black'  : black_bear_imgs,
    'teddy'  : teddy_bear_imgs,
}
path = fastai.Path('bears')
if not path.exists():
    path.mkdir()
for (bear_type, bear_imgs) in bear_types.items():
    print(f'Downloading {len(bear_imgs)} images for {bear_type}')
    dest = (path/bear_type)
    dest.mkdir(exist_ok=True)
    for i, img_url in enumerate(bear_imgs):
        fastbook.download_url(img_url, dest/f'{bear_type}_{i}.jpg')
```

```python
bears = fastai.DataBlock(
 blocks=(fastai.ImageBlock, fastai.CategoryBlock),
 get_items=fastai.get_image_files,
 splitter=fastai.RandomSplitter(valid_pct=0.2, seed=42),
 get_y=fastai.parent_label,
 item_tfms=fastai.Resize(128))
```

```python
bears = bears.new(
 item_tfms=fastai.RandomResizedCrop(224, min_scale=0.5),
 batch_tfms=fastai.aug_transforms())
dls = bears.dataloaders(path, bs=10)
```

```python
learn = fastai.cnn_learner(dls, fastai.resnet18, metrics=fastai.error_rate)
learn.fine_tune(4)
```

```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
lbl_pred = widgets.Label()
out_pl = widgets.Output()
def on_click_classify(change):
 img = fastai.PILImage.create(btn_upload.data[-1])
 out_pl.clear_output()
 with out_pl: display(img.to_thumb(128,128))
 pred,pred_idx,probs = learn.predict(img)
 lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
btn_run.on_click(on_click_classify)
```

```python
display(widgets.VBox([widgets.Label('Select your bear!'), btn_upload, btn_run, out_pl, lbl_pred]))
```
